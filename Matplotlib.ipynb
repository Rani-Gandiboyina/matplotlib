```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
%matplotlib inline
```

```python
x=(-3,5,7)
```

```python
y=(10,2,5)
```

```python
fig=plt.figure(figsize=(15,3))
```

```python
plt.plot(x,y)

plt.xlim(0,10)

plt.ylim(-3,8)

plt.xlabel('X-axis')

plt.ylabel('Y-axis')

plt.title('line plot')

plt.suptitle('Scale comparison',size=20,y=1.03)
```
